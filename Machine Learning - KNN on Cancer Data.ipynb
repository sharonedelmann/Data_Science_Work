# Machine Learning on Cancer Data using KNN

* Using the Breast Cancer Wisconsin (Diagnostic) Database to create a classifier that can help diagnose patients. 

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

#print(cancer.DESCR)
```

The object returned by `load_breast_cancer()` is a scikit-learn Bunch object, which is similar to a dictionary.

```python
cancer.keys()
```

### Step 1

Converting the sklearn.dataset `cancer` to a DataFrame to make things easier. 

```python
def step_one():
    cancer = load_breast_cancer()
    df = pd.DataFrame(cancer.data, columns=cancer.feature_names, index=pd.RangeIndex(start=0, stop=569, step=1))
    df['target'] = cancer.target
    return df

step_one()
```

### Step 2
Determining the class distribution -- how many instances of `malignant` (encoded 0) and how many `benign` (encoded 1)?

```python
def step_two():
    cancerdf = step_one()  
    array_mb_count = np.array([np.sum(cancerdf['target'] == 0),np.sum(cancerdf['target'] > 0)])
    target = pd.Series(data=array_mb_count, index = ['malignant', 'benign'])
    return target

step_two()
```

### Step 3
Splitting the DataFrame into `X` (the data) and `y` (the labels).

```python
def step_three():
    cancerdf = step_one()
    X = cancerdf.drop(['target'], axis=1)
    y = cancerdf['target']
    return X, y

step_three()
```

### Step 4
Using `train_test_split`, split `X` and `y` into training and test sets `(X_train, X_test, y_train, and y_test)`.

```python
from sklearn.model_selection import train_test_split

def step_four():
    X, y = step_three()
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    return X_train, X_test, y_train, y_test

step_four()
```

### Step 5
Using KNeighborsClassifier to fit a k-nearest neighbors (knn) classifier with `X_train`, `y_train` and using one nearest neighbor (`n_neighbors = 1`).

```python
from sklearn.neighbors import KNeighborsClassifier

def step_five():
    X_train, X_test, y_train, y_test = step_four()
    knn = KNeighborsClassifier(n_neighbors = 1)
    knn.fit(X_train, y_train)
    return knn

step_five()
```

### Step 6
Using the above knn classifier, predict the class label using the mean value for each feature.

```python
def step_six():
    cancerdf = step_one()
    means = cancerdf.mean()[:-1].values.reshape(1, -1)
    knn = step_five()
    my_prediction = knn.predict(means)    
    return my_prediction

step_six()
```

### Step 7
Using above knn classifier, predict the class labels for the test set `X_test`.

```python
def step_seven():
    X_train, X_test, y_train, y_test = step_four()
    knn = step_five()
    X_test_prediction = knn.predict(X_test)
    return X_test_prediction

step_seven()
```

### Step 8
Finding the score (mean accuracy) of my knn classifier using `X_test` and `y_test`.

```python
def step_eight():
    X_train, X_test, y_train, y_test = step_four()
    knn = step_five()
    return knn.score(X_test, y_test)

step_eight()
```

### Plot Prediction Scores

Using the plotting function to visualize the differet predicition scores between training and test sets, as well as malignant and benign cells.

```python
def accuracy_plot():
    import matplotlib.pyplot as plt

    %matplotlib notebook

    X_train, X_test, y_train, y_test = step_four()

    # Find the training and testing accuracies by target value (i.e. malignant, benign)
    mal_train_X = X_train[y_train==0]
    mal_train_y = y_train[y_train==0]
    ben_train_X = X_train[y_train==1]
    ben_train_y = y_train[y_train==1]

    mal_test_X = X_test[y_test==0]
    mal_test_y = y_test[y_test==0]
    ben_test_X = X_test[y_test==1]
    ben_test_y = y_test[y_test==1]

    knn = step_five()

    scores = [knn.score(mal_train_X, mal_train_y), knn.score(ben_train_X, ben_train_y), 
              knn.score(mal_test_X, mal_test_y), knn.score(ben_test_X, ben_test_y)]


    plt.figure()

    # Plot the scores as a bar chart
    bars = plt.bar(np.arange(4), scores, color=['#4c72b0','#4c72b0','#55a868','#55a868'])

    # directly label the score onto the bars
    for bar in bars:
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), 
                     ha='center', color='w', fontsize=11)

    # remove all the ticks (both axes), and tick labels on the Y axis
    plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

    # remove the frame of the chart
    for spine in plt.gca().spines.values():
        spine.set_visible(False)

    plt.xticks([0,1,2,3], ['Malignant\nTraining', 'Benign\nTraining', 'Malignant\nTest', 'Benign\nTest'], alpha=0.8);
    plt.title('Training and Test Accuracies for Malignant and Benign Cells', alpha=0.8)

accuracy_plot()
```
