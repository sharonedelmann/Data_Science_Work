#  Understanding Distributions Through Sampling

* Created a number of histogram subplots using the `pyplot subplots` functionality.
* Created an animation, pulling between 100 and 1000 samples from each of the random variables for each plot 

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from matplotlib.widgets import RadioButtons

%matplotlib notebook

# Create the figure and 4 subplots, and create array of the subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

# Set titles for the 4 subplots
axs[0].set_title('Normal')
axs[1].set_title('Gamma')
axs[2].set_title('Exponential')
axs[3].set_title('Uniform')

# Animate
def sharonUpdate(curr):
    # Check if animation is at the last frame, and if so, stop the animation
    if curr == 8: 
        a.event_source.stop()
    num_bins = 10**(2)
    # For the first time through each of the 4, use small sample size
    if curr < 5:
        sample_size = 10**(2)
    # For the second time through each of the 4, use large sample size
    else:
        sample_size = 10**(3)
        
    # Depending on frame/iteration, create the numpy array with the sample size defined above,
    # then create its histogram on the corresponding subplot 
    if curr == 1 or curr == 5:
        x1 = np.random.normal(0, 5, sample_size)
        axs[0].hist(x1, bins=num_bins, alpha=0.5, color='#003f5c')
    if curr == 2 or curr == 6: 
        x2 = np.random.gamma(2, 1.5, sample_size)
        axs[1].hist(x2, bins=num_bins, alpha=0.5, color='#7a5195')
    if curr == 3 or curr == 7: 
        x3 = np.random.exponential(2, sample_size)
        axs[2].hist(x3, bins=num_bins, alpha=0.5, color='#ef5675')
    if curr == 4 or curr == 8: 
        x4 = np.random.uniform(-10, 10, sample_size)
        axs[3].hist(x4, bins=num_bins, alpha=0.5, color='#ffa600')

# Call Animation function at a particular interval
a = animation.FuncAnimation(fig, sharonUpdate, interval=1000, frames=9)

# Show a title for the entire figure
fig.suptitle("Fave Distributions grow from 100 to 1000 Samples\n", fontsize=16)

# Show the figure
plt.show()
```
